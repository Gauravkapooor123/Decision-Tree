```python
df.head(5)
```

```python
# we did this because we have to see that the value will increase on 1.00000 something.
# if it is  or it is not then how the graph will come.


x_grid=np.arange(min(x),max(x),0.0001)
x_grid=x_grid.reshape(len(x_grid),1)
x_grid
```

```python
plt.scatter(x,y,color='r')
plt.plot(x_grid,reg.predict(x_grid))
```

```python
reg.score(x,y)
```
